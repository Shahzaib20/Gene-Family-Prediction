```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Assuming your data is in an Excel file, replace 'your_data.xlsx' with the actual file path
# Adjust the sheet name if your data is in a specific sheet
file_path = '/content/reads.xlsx'
sheet_name = 'Sheet1'  # Change to the actual sheet name

# Read the data from the Excel file
data_df = pd.read_excel(file_path, sheet_name=sheet_name)

# Assuming the second column contains the number of pre-QC reads in million
preqc_counts = data_df.iloc[:, 1].astype(float)

# Assuming the third column contains the number of post-QC reads in million
postqc_counts = data_df.iloc[:, 2].astype(float)

# Assuming the fourth column contains the number of post-hg reads in million
posthg_counts = data_df.iloc[:, 3].astype(float)

# Set the figure size to accommodate the large number of samples
plt.figure(figsize=(12, 6))

# Plotting the histogram for pre-QC reads
sns.histplot(preqc_counts, bins=20, color='#3498db', alpha=0.7)
plt.xlabel('Number of Reads (Million)')
plt.ylabel('Frequency')
plt.title('Pre-QC Read Counts Histogram')
plt.tight_layout()
plt.grid(False)  # Explicitly set grid to False
plt.show()

# Plotting the histogram for post-QC reads
plt.figure(figsize=(12, 6))
sns.histplot(postqc_counts, bins=20, color='#2ecc71', alpha=0.7)
plt.xlabel('Number of Reads (Million)')
plt.ylabel('Frequency')
plt.title('Post-QC Read Counts Histogram')
plt.tight_layout()
plt.grid(False)  # Explicitly set grid to False
plt.show()

# Plotting the histogram for post-hg reads
plt.figure(figsize=(12, 6))
sns.histplot(posthg_counts, bins=20, color='#e74c3c', alpha=0.7)
plt.xlabel('Number of Reads (Million)')
plt.ylabel('Frequency')
plt.title('Post-hg Read Counts Histogram')
plt.tight_layout()
plt.grid(False)  # Explicitly set grid to False
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Assuming your data is in an Excel file, replace 'your_data.xlsx' with the actual file path
# Adjust the sheet name if your data is in a specific sheet
file_path = '/content/reads2.xlsx'
sheet_name = 'Sheet1'  # Change to the actual sheet name

# Read the data from the Excel file
data_df = pd.read_excel(file_path, sheet_name=sheet_name)

# Extract sample names from the first column
sample_names = data_df.iloc[:, 0]

# Assuming the second column contains the number of contigs
contig_counts = data_df.iloc[:, 1].astype(float)

# Assuming the third column contains the size of the largest contig
largest_contig_sizes = data_df.iloc[:, 2].astype(float)

# Assuming the fourth column contains the N50 contig sizes
n50_contig_sizes = data_df.iloc[:, 3].astype(float)

# Convert units for better readability
contig_counts /= 10000  # Convert to 10k units
largest_contig_sizes /= 100000  # Convert to 100k units
n50_contig_sizes /= 5000  # Convert to 5k units

# Set the figure size to accommodate the large number of samples
plt.figure(figsize=(15, 6))

# Set seaborn style for improved aesthetics
sns.set(style="whitegrid")

# Plotting the histogram for the number of contigs
plt.subplot(131)
sns.histplot(contig_counts, bins=20, color='#3498db', alpha=0.7)
plt.xlabel('Number of Contigs (10k units)')
plt.ylabel('Frequency')
plt.title('Number of Contigs Histogram')
plt.grid(False)

# Plotting the histogram for the largest contig sizes
plt.subplot(132)
sns.histplot(largest_contig_sizes, bins=20, color='#2ecc71', alpha=0.7)
plt.xlabel('Largest Contig Size (100k units)')
plt.ylabel('Frequency')
plt.title('Largest Contig Size Histogram')
plt.grid(False)

# Plotting the histogram for N50 contig sizes
plt.subplot(133)
sns.histplot(n50_contig_sizes, bins=20, color='#e74c3c', alpha=0.7)
plt.xlabel('N50 Contig Size (5k units)')
plt.ylabel('Frequency')
plt.title('N50 Contig Size Histogram')
plt.grid(False)

plt.tight_layout()
plt.show()
```
